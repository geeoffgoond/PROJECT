For this project I chose to look at from the 2015 [Puget Sound Regional Travel Study](https://www.psrc.org/travel-surveys-2015-household-survey). Though this study is primarily focused on learning about travel patterns of households within the Puget Sound Region, the survey it uses also asks respondents to rate 9 different "residential factors" (reasons why they choose to live where they live) each on a 1-5 scale. These variables and respondent location (as indicated by zip code) are what I chose to explore.  

I am somewhat familiar with this data-- I currently intern for the Seattle Office of Planning and Community Development, and we are in the process of adding a question to the next iteration of the survey asking why people who moved with in the last five years chose to do so (rather than why they chose the place they moved to), in order to gain a greater understanding of the causes of displacement in Seattle and the Puget Sound Region at large. That said, prior to this project I had not done any analysis to the data.

# Step 1: Get the data <a id="step-1"></a>

We'll specify where the data is on GitHub and then read it with pandas.

```python
#specifying data location data from gitHub, preparing data
import os #os let's us join filepaths
folderData="https://rawgit.com/geeoffgoond/project/master/" #specifying the folder with the data (this is used throughout)
fileXL=os.path.join(folderData,"2015-pr1-hhsurvey-household.xlsx") #specifying the filename

# importing data into pandas
import pandas as pd

psrchh=pd.read_excel(fileXL) #titled 'psrchh' because it's the Puget Sound Regional Council Household Data
```

# Step 2: Limit the data to what is needed <a id="step-1"></a>

Since we're interested in the residential factors and the home zipcode of the respondents, we'll limit the data to those variables, as well the unique identifier for each entry.

```python
#checking what columns we have 
psrchh.columns #the codebook with variable information is in the project folder
```

```python
#we're only interested in a few of those, so let's limit it to those
psrchh=psrchh[["hhid", "h_zip", "h_city", "res_factors_hhchange", "res_factors_afford", "res_factors_school", "res_factors_walk", "res_factors_space", "res_factors_closefam", "res_factors_transit", "res_factors_hwy", "res_factors_30min"]]

#checking to make sure it worked:
psrchh.columns
```

This contains information for all of Washington, but we're only interested in the Puget Sound Region so let's limit it to the counties here:

```python
fileZips=os.path.join(folderData,"waStateZips.xlsx") #getting an excel file with WA state zips by county
waStateZips=pd.read_excel(fileZips) 
#setting the file to the counties we're interested in:
PugetSoundCounties = ["KING", "KITSAP", "PIERCE", "SNOHOMISH", "ISLANDS"] #can change
psZips = waStateZips[(waStateZips['County Name']).isin(PugetSoundCounties)] #creates a dataFrame of data from those counties
#psZips #see the dataFrame to verify (if we'd like)

psZips.reset_index(inplace=True,drop=True) # resetting the index
psrchh = psrchh[(psrchh["h_zip"]).isin(psZips.zip)] #restricting our dataset to just those zipcodes in the counties we specified


psrchh.head() #checking to make sure it still looks good! it does

```

# Step 4: Attempt to cluster the data <a id="step-4"></a>

Ideally we'd be able to cluster the factors if they are highly related to one another, so we'll try to do that here.

```python
len(psrchh.hhid.unique())==len(psrchh.hhid) #checking to make sure hhid (the clustering units) are unique
```

```python
psrchh.index=psrchh.hhid #setting hhid as the index
psrchh.iloc[:,5:14].dtypes #checking dtypes of clustering columns
```

```python
# standardize data:
from sklearn import preprocessing

ps_scaled = preprocessing.scale(psrchh.loc[:,'res_factors_hhchange': 'res_factors_30min'])

#compute similarity matrix
from scipy.spatial.distance import pdist, squareform

psSimi_simple = pdist(ps_scaled,metric='euclidean') # output is not matrix
psSimi_matrix =squareform(psSimi_simple) # this is a matrix
```

```python
from scipy.cluster.hierarchy import linkage, cut_tree
# computing the linkage
psSimi_link = linkage(psSimi_simple,method='average') 
```

We'll graphically cluster the data to try to figure out an appropriate number of clusters:

```python
%matplotlib inline
import seaborn as sns

g = sns.clustermap(psrchh.loc[:,'res_factors_hhchange': 'res_factors_30min'], #plotting the dendogram
                   metric='euclidean',
                   method='average',
                   col_cluster=False,
                  figsize=(10, 20))
```

Based on the dendogram it doesn't look like there's a great way to cluster, but we'll also check numerically:

```python
from sklearn.metrics import silhouette_score
from numpy import ravel

for nOfClusters in range(2,5):
    # get cluster labels
    clusterLabels = ravel(cut_tree(psSimi_link, nOfClusters)) # formatting cut_tree output with ravel
    # get silhouettes
    silhouette_avg = silhouette_score(psSimi_matrix,clusterLabels,metric='precomputed')
    
    # print results:
    print("For n_clusters =", nOfClusters, 
          "The average silhouette_score is:", silhouette_avg)
```

All of those silhouette scores are quite low so we won't cluster the data this way.

# Step 5: Group the data by zipcode: <a id="step-5"></a>

We'll use groupby to create a new dataframe that groups our data by the zipcode of the resident, using the means for each variable by zip:

```python
#we'll also use this data for a map
psMove = psrchh.groupby(['h_zip'], as_index=False)[resfacNames].mean()
psMove #looks great
```

Now we're going to create a column in that df that lists whichever factor was the most impactful in moving for each row:

```python
psMove['factor1'] = psMove[['res_factors_hhchange',
 'res_factors_afford',
 'res_factors_school',
 'res_factors_walk',
 'res_factors_space',
 'res_factors_closefam',
 'res_factors_transit',
 'res_factors_hwy',
 'res_factors_30min']].idxmax(axis=1) #idmax identifies the maximum value in each row and returns the name of the colmun it is in

psMove.factor1 #let's verify that worked
```

```python
psMove.factor1.dtype #checking the dtype
```

```python
#we want that to be a category, so:
from pandas.api.types import CategoricalDtype
psMove.factor1=psMove.factor1.astype(CategoricalDtype(categories=resfacNames, 
                                         ordered=True))
```

```python
psMove.factor1.dtype
```

Let's see which factor is most commonly the highest for a zipcode:

```python
psMove.factor1.describe()
```

We can see that affordability is the top factor in 108 of 160 zipcodes. Let's see what that looks like on a map:

# Step 6: Map the data <a id="step-6"></a>

Let's start by getting the map info:

```python
compressedMap=os.path.join(folderData,'WaZipsMap.zip') #setting the map filepath

import geopandas as gpd

waZipMap = gpd.GeoDataFrame.from_file(compressedMap)

waZipMap.head()
```

Let's make sure our map and data can work together

```python
#checking datatypes
waZipMap.ZCTA5CE10.dtype #map zipcodes
```

```python
psMove.h_zip.dtype #dataset zipcodes
```

```python
waZipMap.ZCTA5CE10=waZipMap.ZCTA5CE10.astype(int) #setting to int
```

Let's restrict map to just those counties we specified earlier

```python
#psZips.zip.astype(int) #making sure it's the same type of data
waZipMap=waZipMap[(waZipMap.ZCTA5CE10.isin(psZips.zip))]
waZipMap
```

```python
psMoveMap=waZipMap.merge(psMove, left_on='ZCTA5CE10', right_on='h_zip') ## merging data into shapefile
```

Now we'll plot a map showing the most greatest move factor for each zipcode.

```python
%matplotlib inline
base = waZipMap.plot(color='#ededed',figsize=(20,13)) #setting specifications for the map

psMoveMap.plot(ax=base,column='factor1',cmap='rainbow',categorical=True,legend=True) #plotting factor1
```

Since affordability is most commonly the highest factor, let's map each zipcode's mean affordability residential factor score as well: 

```python
%matplotlib inline
base = waZipMap.plot(color='grey',figsize=(20,13))

#psrchhWAmap.plot(ax=base, color='pink')
psMoveMap.plot(ax=base,column='res_factors_afford', cmap='YlGnBu', scheme='Quantiles', k=5, legend=True) #affordability
```
